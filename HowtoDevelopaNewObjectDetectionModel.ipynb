```python
# example of extracting bounding boxes from an annotation file
from xml.etree import ElementTree
# function to extract bounding boxes from an annotation file
def extract_boxes(filename):
    # load and parse the file
    tree = ElementTree.parse(filename)
    # get the root of the document
    root = tree.getroot()
    # extract each bounding box
    boxes = list()
    for box in root.findall('.//bndbox'):
        xmin = int(box.find('xmin').text)
        ymin = int(box.find('ymin').text)
        xmax = int(box.find('xmax').text)
        ymax = int(box.find('ymax').text)
        coors = [xmin, ymin, xmax, ymax]
        boxes.append(coors)
        # extract image dimensions
        width = int(root.find('.//size/width').text)
        height = int(root.find('.//size/height').text)
    return boxes, width, height

# extract details form annotation file
boxes, w, h = extract_boxes('kangaroo/annots/00001.xml')
# summarize extracted details
print(boxes, w, h)
```

```python
# Importar librerías necesarias
import cv2
import os
import numpy as np
from ultralytics import YOLO
from matplotlib import pyplot as plt

# Clase para manejar el dataset de canguros
class KangarooDatasetYOLO:
    def __init__(self, dataset_dir, is_train=True):
        self.dataset_dir = dataset_dir
        self.is_train = is_train
        self.images = []
        self.annotations = []

        # Cargar imágenes y anotaciones
        self.load_dataset()

    def load_dataset(self):
        images_dir = f"{self.dataset_dir}/images/"
        annotations_dir = f"{self.dataset_dir}/annots/"

        for image_name in sorted(os.listdir(images_dir)):
            image_id = image_name[:-4]  # Eliminar la extensión
            if image_id in ['00090']:  # Omitir imágenes malas
                continue
            if self.is_train and int(image_id) >= 150:
                continue
            if not self.is_train and int(image_id) < 150:
                continue

            image_path = images_dir + image_name
            annotation_path = annotations_dir + image_id + '.xml'

            self.images.append(image_path)
            self.annotations.append(annotation_path)

    def extract_boxes(self, annotation_path):
        from xml.etree import ElementTree as ET

        tree = ET.parse(annotation_path)
        root = tree.getroot()
        boxes = []

        for box in root.findall('.//bndbox'):
            xmin = int(box.find('xmin').text)
            ymin = int(box.find('ymin').text)
            xmax = int(box.find('xmax').text)
            ymax = int(box.find('ymax').text)
            boxes.append([xmin, ymin, xmax, ymax])

        width = int(root.find('.//size/width').text)
        height = int(root.find('.//size/height').text)
        return boxes, width, height

# Función para visualizar imágenes con bounding boxes
def visualize_image(image_path, boxes):
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    for box in boxes:
        xmin, ymin, xmax, ymax = box
        cv2.rectangle(image, (xmin, ymin), (xmax, ymax), (255, 0, 0), 2)

    plt.imshow(image)
    plt.axis('off')
    plt.show()

# Entrenar un modelo YOLOv8 en el dataset
def train_yolo_model(dataset_path):
    # Crear el modelo YOLOv8
    model = YOLO('yolov8n.yaml')  # Modelo "nano" de YOLOv8 para entrenamiento rápido

    # Preparar configuración de entrenamiento
    model.train(
        data='C:/Users/RSSpe/Pruebas_IA/DeepLearningForCV/yolo_dataset.yaml',  # Archivo de configuración con clases y datos
        epochs=50,
        imgsz=640,
        batch=16,
        name='kangaroo_detection'
    )
    return model

# Detectar y visualizar resultados
def detect_with_yolo(model_path, image_path):
    model = YOLO(model_path)
    results = model(image_path)

    # Extraer y mostrar detecciones
    result_image = results[0].plot()
    plt.imshow(result_image)
    plt.axis('off')
    plt.show()

# Ejemplo de uso
if __name__ == "__main__":
    dataset_path = "kangaroo"

    # Preparar dataset
    dataset = KangarooDatasetYOLO(dataset_path, is_train=True)

    # Visualizar un ejemplo
    image_path = dataset.images[0]
    boxes, _, _ = dataset.extract_boxes(dataset.annotations[0])
    visualize_image(image_path, boxes)

    # Entrenar YOLOv8
    model = train_yolo_model(dataset_path)
```

```python
from ultralytics import YOLO
import os
import xml.etree.ElementTree as ET
import shutil
import yaml

# Función para convertir anotaciones XML a formato YOLO
def convert_xml_to_yolo(xml_file, output_file, class_name_map, image_width, image_height):
    tree = ET.parse(xml_file)
    root = tree.getroot()

    with open(output_file, 'w') as f:
        for obj in root.findall('object'):
            class_name = obj.find('name').text
            if class_name not in class_name_map:
                continue
            class_id = class_name_map[class_name]
            
            bndbox = obj.find('bndbox')
            xmin = int(bndbox.find('xmin').text)
            ymin = int(bndbox.find('ymin').text)
            xmax = int(bndbox.find('xmax').text)
            ymax = int(bndbox.find('ymax').text)

            # Convert bounding box to YOLO format
            x_center = ((xmin + xmax) / 2) / image_width
            y_center = ((ymin + ymax) / 2) / image_height
            bbox_width = (xmax - xmin) / image_width
            bbox_height = (ymax - ymin) / image_height

            f.write(f"{class_id} {x_center} {y_center} {bbox_width} {bbox_height}\n")

# Preparar dataset para YOLO
def prepare_yolo_dataset(dataset_dir, output_dir, class_names):
    images_dir = os.path.join(dataset_dir, 'images')
    annotations_dir = os.path.join(dataset_dir, 'annots')
    
    train_images_dir = os.path.join(output_dir, 'images/train')
    val_images_dir = os.path.join(output_dir, 'images/val')
    train_labels_dir = os.path.join(output_dir, 'labels/train')
    val_labels_dir = os.path.join(output_dir, 'labels/val')

    # Crear carpetas
    for folder in [train_images_dir, val_images_dir, train_labels_dir, val_labels_dir]:
        os.makedirs(folder, exist_ok=True)
    
    # Crear el mapeo de clases
    class_name_map = {name: idx for idx, name in enumerate(class_names)}
    
    for split in ['train', 'val']:
        is_train = (split == 'train')
        output_images_dir = train_images_dir if is_train else val_images_dir
        output_labels_dir = train_labels_dir if is_train else val_labels_dir

        for filename in os.listdir(images_dir):
            image_id = filename[:-4]
            img_path = os.path.join(images_dir, filename)
            ann_path = os.path.join(annotations_dir, image_id + '.xml')
            
            if not os.path.exists(ann_path):
                continue

            # Condición para separar train y val
            if is_train and int(image_id) >= 150:
                continue
            if not is_train and int(image_id) < 150:
                continue
            
            # Copiar imagen
            shutil.copy(img_path, output_images_dir)

            # Convertir XML a formato YOLO
            tree = ET.parse(ann_path)
            root = tree.getroot()
            width = int(root.find('.//size/width').text)
            height = int(root.find('.//size/height').text)
            
            label_file = os.path.join(output_labels_dir, image_id + '.txt')
            convert_xml_to_yolo(ann_path, label_file, class_name_map, width, height)
    
    # Crear archivo YAML para YOLOv8
    yaml_data = {
        'train': os.path.join(output_dir, 'images/train'),
        'val': os.path.join(output_dir, 'images/val'),
        'nc': len(class_names),
        'names': class_names
    }
    
    with open(os.path.join(output_dir, 'data.yaml'), 'w') as yaml_file:
        yaml.dump(yaml_data, yaml_file)

# Entrenar el modelo YOLOv8
def train_yolo_model(data_yaml, model='yolov8n', epochs=5, imgsz=640, batch=16):
    model = YOLO(model + '.yaml')  # Cargar modelo YOLOv8
    model.train(
        data=data_yaml,
        epochs=epochs,
        imgsz=imgsz,
        batch=batch,
        name='kangaroo_detection'
    )

# Ejemplo de uso
if __name__ == "__main__":
    dataset_dir = "kangaroo"  # Directorio donde están las imágenes y anotaciones XML
    output_dir = "yolo_kangaroo"  # Directorio donde se guardará el dataset formateado para YOLO
    class_names = ["kangaroo"]  # Nombres de las clases
    
    # Preparar el dataset
    prepare_yolo_dataset(dataset_dir, output_dir, class_names)
    
    # Entrenar YOLOv8
    data_yaml_path = os.path.join(output_dir, 'data.yaml')
    train_yolo_model(data_yaml_path, model='yolov8n', epochs=5)
```
